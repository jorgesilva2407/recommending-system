```python
import pandas as pd
from fpgrowth_py import fpgrowth
```

```python
df = pd.read_csv("./2023_spotify_ds2.csv")
df
```

```python
# playlists = df.groupby("pid")["track_uri"].apply(list).to_list()
playlists = (
    df.groupby("pid")[["artist_name", "track_name"]]
    .apply(lambda x: x.to_records(index=False).tolist())
    .to_list()
)
```

```python
_, rules = fpgrowth(playlists, minSupRatio=0.1, minConf=0.5)
[(rule[0], rule[1]) for rule in rules]
```
